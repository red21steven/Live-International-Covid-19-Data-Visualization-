# In this project I've done some analysis and visualization on COVID-19 with the data available at [Worldometers.](https://www.worldometers.info/coronavirus/) 

**In this project I've built a web scraper to scrape the data from the public domain and made some analysis & visualization with the scraped data.**

**This project gives the audience a live visual on the data from the website whenever it's updated all need to do is just run each and every cell to get the new visual on the available data in an instant.**

***This project is a work in progress but all the visual given in this workspace is constantly checked and updated inorder to give the accurate visual analysis.***



```python
#Imported the necessary packages and libraries
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import plotly.express as px
import pandas as pd
import requests
import csv

#Building a Webscraper
request = requests.get('https://www.worldometers.info/coronavirus/')

soup = BeautifulSoup(request.text, 'html5lib')

world = soup.select_one('.table')

table_r = world.find_all('tr')[9:230]

csv_data = [['country name', 'Total cases', 'New cases', 'Total death', 'New death', 'Total recovered', 'Active cases', 'Critical cases', 'total_test', 'total_population']]


for r in table_r:
  country_n = r.find_all('td')[1].text
  total_cases = r.find_all('td')[2].text
  new_cases = r.find_all('td')[3].text
  total_death = r.find_all('td')[4].text
  new_death = r.find_all('td')[5].text
  total_recovered = r.find_all('td')[6].text
  Active_cases = r.find_all('td')[8].text
  critical_cases = r.find_all('td')[9].text
  total_test = r.find_all('td')[12].text
  total_population = r.find_all('td')[14].text
  csv_data.append([country_n, total_cases, new_cases, total_death, new_death, total_recovered, Active_cases, critical_cases, total_test, total_population])
print(csv_data)

#Save the scraped data in an excel sheet
with open('covid cases worldwide.csv', 'w') as file:
  write = csv.writer(file)
  write.writerows(csv_data)

#Function to read the CSV file
def read_file(paths, filename):
  return pd.read_csv(path +'/'+ filename)

path = '/content'
covid_data = read_file(path, 'covid cases worldwide.csv')
```
